```python
import numpy as np
def calc_std_oneEvent(wfs):
    stds = []
    for wf in wfs:
        std = np.std(wf)
        stds.append(std)
    stds = np.array(stds)
    return stds

```

```python
def collect_delta_response_templates(df, evtno, chaid, neg_thre=-1000):
    full_wfs = []
    for evtid in range(evtno):
        wfs = list(df['Data'].iloc[evtid][chaid])
        full_wfs = full_wfs + wfs
        
    stencil_length = 10
    responses = []
    for i in range(len(full_wfs)-11):
        if full_wfs[i+1] < neg_thre and full_wfs[i] >= neg_thre:
            one_stencil = full_wfs[i-3:i+7]
            responses.append(one_stencil)
    return responses
```
